```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
np.load(r'WWW.npy').shape
```

```python
stim = np.unpackbits(np.load(r'WWW.npy'), axis = 2)
stim.shape

stim[283]
```

stim[f, r, c] gives you the pixel value (0 or 1) of frame f, row r, column c.

There are 624 frames, which were shown at 1440 Hz. The duration of each frame was therefore about 694 Âµs, and the total duration about 433 ms.

Each frame is 256x256 pixels. Don't be surpised to find a reguler grid of on pixels (every 4th pixel) outside the actual stimulus. We did this to have an approximately gray background, in order to avoid large changes in contrast during stimulus onset and offset. 

```python
fig, axes = plt.subplots(6, 8, figsize = (8, 6))
frame = 283
for y in range(6):
    for x in range(8):
        ax = axes[y, x]
        ax.imshow(stim[frame])
        ax.axis('off')
        ax.text(10, 10, frame, c = 'white', va = 'top', size = 6)
        frame += 1
```
